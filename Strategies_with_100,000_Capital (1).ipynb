#Momentum Asset Allocation Strategy

```python
df = pd.read_csv('/content/drive/MyDrive/CSE All-Share Historical Data.csv')
```

```python
df["Price"] = df["Price"].str.replace(",", "").astype(float)
```

```python
df['SMA'] = ta.SMA(df['Price'], timeperiod=20)  # Simple Moving Average
df['RSI'] = ta.RSI(df['Price'], timeperiod=14)  # Relative Strength Index
```

```python
def get_signal(row):
    if row['Price'] > row['SMA'] and row['RSI'] > 70:
        return 'Sell'
    elif row['Price'] < row['SMA'] and row['RSI'] < 30:
        return 'Buy'
    else:
        return 'Hold'
```

```python
df['Signal'] = df.apply(get_signal, axis=1)
```

```python
allocation_percentage = 0.5
df['Allocation'] = allocation_percentage
```

```python
df['Capital'] = df['Allocation'] * 100000
```

```python
for i in range(len(df)):
    if i % 30 == 0:
        df.at[i, 'Capital'] = df.at[i, 'Allocation'] * 100000
```

```python
for i, row in df.iterrows():
    if row['Signal'] == 'Buy':
        shares = row['Capital'] / row['Price']
    elif row['Signal'] == 'Sell':
        df.at[i, 'Capital'] = shares * row['Price']
    else:
        continue

total_profit_loss = df.iloc[-1]['Capital'] - 100000
```

```python
for i, row in df.iterrows():
    print(f"Day {i+1}: {row['Signal']}")
print(f"Total profit or loss: {total_profit_loss}")
```

#Payday Anomaly

```python
df = pd.read_csv('/content/drive/MyDrive/CSE All-Share Historical Data.csv')
```

```python
df["Price"] = df["Price"].str.replace(",", "").astype(float)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['DayOfWeek'] = df['Date'].dt.dayofweek
```

```python
df['SMA'] = ta.SMA(df['Price'], timeperiod=20)  # Simple Moving Average
df['RSI'] = ta.RSI(df['Price'], timeperiod=14)  # Relative Strength Index
```

```python
def get_signal(row):
    if row['DayOfWeek'] == 4:  # If it's Friday
        if row['Price'] > row['SMA'] and row['RSI'] > 70:
            return 'Sell'
        elif row['Price'] < row['SMA'] and row['RSI'] < 30:
            return 'Buy'
        else:
            return 'Hold'
    else:
        return 'Hold'
```

```python
df['Signal'] = df.apply(get_signal, axis=1)
```

```python
df['Capital'] = 100000
```

```python
for i in range(len(df)):
    if df.at[i, 'Date'].day == 1:
        df.at[i, 'Capital'] = 100000
```

```python
shares = 0
for i, row in df.iterrows():
    if row['Signal'] == 'Buy':
        shares = row['Capital'] / row['Price']
    elif row['Signal'] == 'Sell':
        df.at[i, 'Capital'] = shares * row['Price']
    else:
        continue

total_profit_loss = df.iloc[-1]['Capital'] - 100000
```

```python
for i, row in df.iterrows():
    print(f"Day {i+1}: {row['Signal']}")
print(f"Total profit or loss: {total_profit_loss}")
```

#Volatility Risk Premium Effect

```python
df = pd.read_csv('/content/drive/MyDrive/CSE All-Share Historical Data.csv')
```

```python
df["Price"] = df["Price"].str.replace(",", "").astype(float)
df["High"] = df["High"].str.replace(",", "").astype(float)
df["Low"] = df["Low"].str.replace(",", "").astype(float)
```

```python
df['SMA'] = ta.SMA(df['Price'], timeperiod=20)  # Simple Moving Average
df['ATR'] = ta.ATR(df['High'], df['Low'], df['Price'], timeperiod=14)  # Average True Range
```

```python
def get_signal(row, threshold):
    if row['Price'] > row['SMA'] + (threshold * row['ATR']):
        return 'Sell'
    elif row['Price'] < row['SMA'] - (threshold * row['ATR']):
        return 'Buy'
    else:
        return 'Hold'
```

```python
threshold = 1.0
df['Signal'] = df.apply(lambda row: get_signal(row, threshold), axis=1)
```

```python
df['Capital'] = 100000
df['Date'] = pd.to_datetime(df['Date'])
```

```python
for i in range(len(df)):
    if df.at[i, 'Date'].day == 1:
        df.at[i, 'Capital'] = 100000
```

```python
shares = 0
for i, row in df.iterrows():
    if row['Signal'] == 'Buy':
        shares = row['Capital'] / row['Price']
    elif row['Signal'] == 'Sell':
        df.at[i, 'Capital'] = shares * row['Price']
    else:
        continue

total_profit_loss = df.iloc[-1]['Capital'] - 100000
```

```python
for i, row in df.iterrows():
    print(f"Day {i+1}: {row['Signal']}")
print(f"Total profit or loss: {total_profit_loss}")
```
